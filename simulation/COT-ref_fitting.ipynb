```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
from scipy.optimize import curve_fit
```

```python
with h5py.File('data/20181018_central_asia_2_470cloud_test2_20181018/pre-data.h5') as f:
    ref = f['mca_ipa_thick/ref'][...]
    cot = f['mca_ipa_thick/cot'][...]
```

```python
def func(x, a, b, c):
     return a * np.exp(-b * x) + c
    
#(tau + a*x) / (tau + x)
```

```python
def cot_fitting_func(x, a, b, c):
     return (b + (a * x))/(b+x) + c
```

```python
popt, pcov = curve_fit(func, cot, ref)
```

```python
plt.scatter(cot, ref)
plt.plot(cot, func(cot, *popt), 'r')
plt.show()
```
